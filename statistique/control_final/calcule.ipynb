```python
# Exo 1_1 
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Données
nombres_enfants = np.array([0, 1, 2, 3, 4, 5, 6])
effectifs = np.array([18, 32, 66, 41, 32, 9, 2])

# 1. Description de la population
effectif_total = np.sum(effectifs)
unite_statistique = "famille"
caracteres_etudies = "nombre d'enfants à charge"
type_variables = "quantitative discrète"
valeurs_modalites = nombres_enfants

# 2. Caractéristiques de tendance centrale
moyenne = np.mean(effectifs)
mediane = np.median(effectifs)
mode_result = stats.mode(effectifs)
mode = mode_result[0]

# 3. Caractéristiques de dispersion
etendue = np.max(nombres_enfants) - np.min(nombres_enfants)
quartiles = np.percentile(np.repeat(nombres_enfants, effectifs), [25, 75])
ecart_inter_quartile = quartiles[1] - quartiles[0]
variance = np.sum((nombres_enfants - moyenne)**2 * effectifs) / effectif_total
ecart_type = np.sqrt(variance)

# 4. Tracé des graphiques
plt.figure(figsize=(12, 6))

# Histogramme
plt.subplot(1, 3, 1)
plt.bar(nombres_enfants, effectifs, color='skyblue', edgecolor='black')
plt.title("Histogramme")
plt.xlabel("Nombre d'enfants à charge")
plt.ylabel("Effectif")

# Densité
plt.subplot(1, 3, 2)
plt.bar(nombres_enfants, effectifs / effectif_total, color='skyblue', edgecolor='black')
plt.title("Densité")
plt.xlabel("Nombre d'enfants à charge")
plt.ylabel("Fréquence relative")

# Boîte à moustache
plt.subplot(1, 3, 3)
plt.boxplot(np.repeat(nombres_enfants, effectifs), vert=False)
plt.title("Boîte à moustache")
plt.xlabel("Nombre d'enfants à charge")

plt.tight_layout()
plt.show()

# Affichage des résultats
print("1. Description de la population:")
print(f"Effectif total : {effectif_total}")
print(f"Unité statistique : {unite_statistique}")
print(f"Caractères étudiés : {caracteres_etudies}")
print(f"Type des variables : {type_variables}")
print(f"Valeurs ou modalités : {valeurs_modalites}")

print("\n2. Caractéristiques de tendance centrale:")
print(f"Moyenne : {moyenne}")
print(f"Médiane : {mediane}")
print(f"Mode : {mode}")

print("\n3. Caractéristiques de dispersion:")
print(f"Étendue : {etendue}")
print(f"Écart inter-quartile : {ecart_inter_quartile}")
print(f"Variance : {variance}")
print(f"Écart-type : {ecart_type}")
```

```python
# TP 1_5

import numpy as np
import matplotlib.pyplot as plt

# Données
bornes_inf = [1.9, 2.0, 2.1, 2.2, 2.3, 2.4, 2.5]
bornes_sup = [2.0, 2.1, 2.2, 2.3, 2.4, 2.5, 2.6]
effectifs = [3, 9, 18, 29, 25, 10, 6]

# 1. Description de la population
effectif_total = np.sum(effectifs)
unite_statistique = "tube"
caracteres_etudies = "diamètre en décimètre"
type_variables = "quantitative continue"

# 2. Caractéristiques de tendance centrale
valeurs = np.array([(inf + sup) / 2 for inf, sup in zip(bornes_inf, bornes_sup)])

moyenne = np.mean(effectifs)
mediane = np.median(effectifs)
mode_res = stats.mode(effectifs)
mode = mode_res[0]

# 3. Caractéristiques de dispersion (approximatives pour les données groupées)
approximation_variance = np.sum((valeurs_modalites - moyenne)**2 * effectifs) / effectif_total
approximation_ecart_type = np.sqrt(approximation_variance)

# 4. Tracé des graphiques
plt.figure(figsize=(12, 6))

# Histogramme (pour données groupées)
plt.subplot(1, 3, 1)
plt.bar(valeurs_modalites, effectifs, width=0.1, edgecolor='black')
plt.title("Histogramme")
plt.xlabel("Diamètre en décimètre")
plt.ylabel("Effectif")

# Tracé de la densité
plt.subplot(1, 3, 2)
density = effectifs / np.sum(effectifs)
plt.plot(valeurs, density, '-o', color='orange')
plt.xlabel('Diamètre (en décimètre)')
plt.ylabel('Densité')
plt.title('Densité')

# Boîte à moustaches
plt.subplot(1, 3, 3)
plt.boxplot(valeurs_modalites, vert=False)
plt.title("Boîte à moustache")
plt.xlabel("Diamètre en décimètre")

plt.show()

# Affichage des résultats
print("1. Description de la population:")
print(f"Effectif total : {effectif_total}")
print(f"Unité statistique : {unite_statistique}")
print(f"Caractères étudiés : {caracteres_etudies}")
print(f"Type des variables : {type_variables}")
print(f"Valeurs ou modalités : {valeurs_modalites}")

print("\n2. Caractéristiques de tendance centrale:")
print(f"Moyenne : {moyenne}")
print(f"Mediane : {mediane}")
print(f"Mode : {mode}")

print("\n3. Caractéristiques de dispersion (approximatives pour les données groupées):")
print(f"Approximation de la variance : {approximation_variance}")
print(f"Approximation de l'écart-type : {approximation_ecart_type}")
```
